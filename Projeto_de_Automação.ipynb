### segue sequência para abrir o Chrome
sempre lembrar de 1º criar um passo a passo para poder seguir quando for automatizar uma tarefa

```python
pyautogui.PAUSE = 0.5 #só funcionará nos comandos que estiverem com pyautogui
pyautogui.press("win")
#colocar o tempo de execução
pyautogui.write("chrome")
time.sleep(3)
pyautogui.press("enter")
#pyautogui.hotkey("ctrl", "shift", "n")

time.sleep(3)

# > antes de acessar o link eu tenho que escolher a opção de conta
pyautogui.click(x=822, y=623)
#caso o chorme ja esteja aberto eu pedi para ele abrir uma nova aba...
pyautogui.click(x=764, y=23)

# > escrever o link
pyautogui.write("https://dlp.hashtagtreinamentos.com/python/intensivao/login")
pyautogui.press("enter")
# Após apertar enter eu quero que esse código (apenas esse) de um tempo para porcessar os acontecimentos - como carregar a página toda...
# - Então eu vou importar o time para chamar sleep 
time.sleep(5)
# > A diferença entre time.sleep e o pyautogui.PAUSE é que:
# *time.sleep vai funcionar so para aquela chamada
# *pyautogui.PAUSE vai funcionar para cada um dos comando do pyautogui
pyautogui.click(x=781, y=589)
pyautogui.write("rm6@gmail.com")
time.sleep(2)
pyautogui.click(x=715, y=750)
# > temos a opção do pyautogui.press("tab")
pyautogui.write("senha")
pyautogui.press("enter")

import pandas as pd
 
dados = pd.read_csv("produtos.csv")
print(dados)

time.sleep(5)
# > Agora vamos cadastrar os produtos!!!!
linha = 0 
pyautogui.click(x=661, y=417)

# para que eu consiga localizar o produto e escreva no campo selecionado eu uso o seguinte código
codigo = dados.loc[linha, "codigo"] # como identifiquei la em cima que quero a linha 0 como linha , eu chamo entre [linha e a coluna que identifica na minha base de dados]
time.sleep(3)
pyautogui.write(str(codigo)) #usando o str eu me certifico de que ele vai escrever um texto, porque pode acontecer de ter que escrever numeros e o código não prduzir como a gnt precisa

# -->Preenchendo os demais campos
pyautogui.press("tab")
pyautogui.write(str(dados.loc[linha, "marca"]))
pyautogui.press("tab")
pyautogui.write(str(dados.loc[linha, "tipo"]))
pyautogui.press("tab")
pyautogui.write(str(dados.loc[linha, "categoria"]))
pyautogui.press("tab")
pyautogui.write(str(dados.loc[linha, "preco_unitario"]))
pyautogui.press("tab")
pyautogui.write(str(dados.loc[linha, "custo"]))
pyautogui.press("tab")

# UM CASO A OBSERVAR: 
# > A coluna (obs) tem campos vazios então temos que tratar esses campos vazios da seguinte forma
#se for o caso do campos estiver vazio ele simplemente da o tab e não preenche por isso o pyautogui.press("tab") estará fora da identação
#ele so será passado para o campo, se o campos tiver informação e por isso ele esta identado no if para que ele passe a informação para o campo quando for selecionado
obs = dados.loc[linha, "obs"]
if not pd.isna(obs): #isna verifica se esta vazio
    pyautogui.write(str(dados.loc[linha, "obs"]))
pyautogui.press("tab")
pyautogui.press("enter")

# > Para dar seguimento para conseguir preencher todos os dados novamente, la na pagina que foi direcionado vai ser preciso dar um (scroll - rolar a barra para cima), dessa forma ele responderá
pyautogui.scroll(5000)
#para rolar para cima é +(positivo = subir a tela)
#para rolar para baixo é -(negativo = desce a tela)

# > Será feito um for para repetir o cadastro de todos os produtos do arquivo produtos.csv
for linha in dados.index: #quer dizer = para cada linha da minha tabela 
    #executando todo código novamente que esta acima
    pyautogui.click(x=661, y=417)
    codigo = dados.loc[linha, "codigo"]
    time.sleep(3)
    pyautogui.write(str(codigo))

    pyautogui.press("tab")
    pyautogui.write(str(dados.loc[linha, "marca"]))
    pyautogui.press("tab")
    pyautogui.write(str(dados.loc[linha, "tipo"]))
    pyautogui.press("tab")
    pyautogui.write(str(dados.loc[linha, "categoria"]))
    pyautogui.press("tab")
    pyautogui.write(str(dados.loc[linha, "preco_unitario"]))
    pyautogui.press("tab")
    pyautogui.write(str(dados.loc[linha, "custo"]))
    pyautogui.press("tab")


    obs = dados.loc[linha, "obs"]
    if not pd.isna(obs): #isna verifica se esta vazio
        pyautogui.write(str(dados.loc[linha, "obs"]))
    pyautogui.press("tab")
    pyautogui.press("enter")

    pyautogui.scroll(5000)


```
