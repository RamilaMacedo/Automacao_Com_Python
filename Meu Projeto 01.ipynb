```python
import pandas as pd

tabela = pd.read_csv(r"C:\Users\ramil\Intensivão Python\Compras.csv", sep=';') #r significa: ler o arquivo da maneira que ele está
#sep: indicar separador do documento que vai abrir
display(tabela) #mostrar tabela

total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto/quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)
      
    
```

```python
import pyperclip

#entrar no email e enviar o arquivo
pg.hotkey('ctrl', 't') #abrir aba no chrome
pg.write('https://mail.google.com/mail/u/0/#inbox')
pg.press('enter')

time.sleep(5)

#apertar no ícone escrever para mandar email

pg.click(x=113, y=313)

time.sleep(5)

pg.write("atualizadosvendas@gmail.com")
pg.press("tab") #1 TAB  para selecionar o destinatário
pg.press('tab') #2º TAB  para mudar de opção e ir para assunto


pyperclip.copy("Relatório de vendas") # o pyperclip entra no comando pois, como ele tem um caracter de acento, la no email não vai aparecer, então por isso usar o pyperclip para alterar
pg.hotkey('ctrl', 'v')
pg.press("tab") # 3º TAB para ir para o corpo do email


texto = f'''
Valores Extraídos do Relatório de vendas do arquivo CSV

Total de Gastos: R${total_gasto:,.2f}
Quantidade: R${quantidade:,}
Preço Médio: R${preco_medio:,.2f}
'''
#abrir 3 as simples/duplas para escrever código com várias linhas

pyperclip.copy(texto)
pg.hotkey("ctrl", "v")

#enviar email
pg.hotkey("ctrl", "enter")


```
